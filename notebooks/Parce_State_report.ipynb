```python
import pandas as pd
import numpy as np
```

## Initialize files/variables

```python
countries = ['Angola','Benin','Botswana','Burkina Faso','Burundi','Cabo Verde','Cameroon',\
             'Central African Republic','Chad','Comoros','Congo, Democratic Republic of the',\
             'Congo, Republic of the',"Cote d'Ivoire",'Djibouti','Equatorial Guinea','Eritrea',\
             'Eswatini','Ethiopia','Gabon','Gambia, The','Ghana','Guinea','Guinea-Bissau','Kenya',\
             'Lesotho','Liberia','Madagascar','Malawi','Mali','Mauritania','Mauritius','Mozambique',\
             'Namibia','Niger','Nigeria','Rwanda','Sao Tome and Principe','Senegal','Seychelles',\
             'Sierra Leone','Somalia','South Africa','South Sudan','Sudan','Tanzania','Togo','Uganda',\
             'Zambia','Zimbabwe']
```

```python
State_reports = pd.DataFrame(columns = ['CountryName','Summary','Section_1a','Section_1b','Section_1c',\
                                        'Section_1d','Section_1e','Section_1f','Section_1g','Section_2a','Section_2b',\
                                       'Section_2c','Section_2d','Section_3','Section_4','Section_5',\
                                        'Section_6','Section_7a','Section_7b','Section_7c','Section_7d',\
                                       'Section_7e'])
```

## def/classes that will be called later

```python
def strip_section(infile,titles):
    exit_index = len(infile)
    
    for i in range(len(infile)):
        if infile[i] in titles:
            exit_index = i
            break
            
    section = ' '.join(infile[0:exit_index])
        
    return section
```

```python
class Parcedatafile:
    def __init__(self,infile,countries):
            
        titles = ['EXECUTIVE SUMMARY','SECTION 1. RESPECT FOR THE INTEGRITY OF THE PERSON, INCLUDING FREEDOM FROM:',
                  'a. Arbitrary Deprivation of Life and Other Unlawful or Politically Motivated Killings',
                  'b. Disappearance', 'c. Torture and Other Cruel, Inhuman, or Degrading Treatment or Punishment',
                  'd. Arbitrary Arrest or Detention','e. Denial of Fair Public Trial',
                  'f. Arbitrary or Unlawful Interference with Privacy, Family, Home, or Correspondence',
                  'g. Abuses in Internal Conflict','SECTION 2. RESPECT FOR CIVIL LIBERTIES, INCLUDING:',
                  'a. Freedom of Expression, Including for the Press','b. Freedoms of Peaceful Assembly and Association',
                  'c. Freedom of Religion','d. Freedom of Movement','SECTION 3. FREEDOM TO PARTICIPATE IN THE POLITICAL PROCESS',
                  'SECTION 4. CORRUPTION AND LACK OF TRANSPARENCY IN GOVERNMENT',
                  'SECTION 5. GOVERNMENTAL ATTITUDE REGARDING INTERNATIONAL AND NONGOVERNMENTAL INVESTIGATION OF ALLEGED ABUSES OF HUMAN RIGHTS',
                  'SECTION 6. DISCRIMINATION, SOCIETAL ABUSES, AND TRAFFICKING IN PERSONS','SECTION 7. WORKER RIGHTS',
                  'a. Freedom of Association and the Right to Collective Bargaining',
                  'b. Prohibition of Forced or Compulsory Labor','c. Prohibition of Child Labor and Minimum Age for Employment',
                  'd. Discrimination with Respect to Employment and Occupation','e. Acceptable Conditions of Work']
        
        # initialize all of the variables
        self.country, self.summary = np.nan, np.nan
        self.section1a, self.section1b, self.section1c, self.section1d = np.nan, np.nan, np.nan, np.nan
        self.section1e, self.section1f, self.section1g = np.nan, np.nan, np.nan
        self.section2a, self.section2b, self.section2c, self.section2d = np.nan, np.nan, np.nan, np.nan
        self.section3, self.section4, self.section5, self.section6 = np.nan, np.nan, np.nan, np.nan
        self.section7a, self.section7b, self.section7c, self.section7d = np.nan, np.nan, np.nan, np.nan
        self.section7e = np.nan
        
        for i in range(len(infile)-1):
            
            if infile[i] in countries:
                self.country = infile[i]
            if infile[i] == titles[0]: 
                self.summary = strip_section(infile[i+1:],titles)
            # SECTION 1. RESPECT FOR THE INTEGRITY OF THE PERSON
            if infile[i] == titles[2]:
                self.section1a = strip_section(infile[i+1:],titles)
            if infile[i] == titles[3]:
                self.section1b = strip_section(infile[i+1:],titles)
            if infile[i] == titles[4]:
                self.section1c = strip_section(infile[i+1:],titles)
            if infile[i] == titles[5]:
                self.section1d = strip_section(infile[i+1:],titles)
            if infile[i] == titles[6]:
                self.section1e = strip_section(infile[i+1:],titles)
            if infile[i] == titles[7]:
                self.section1f = strip_section(infile[i+1:],titles)
            if infile[i] == titles[8]:
                self.section1g = strip_section(infile[i+1:],titles)
            # SECTION 2. RESPECT FOR CIVIL LIBERTIES
            if infile[i] == titles[10]:
                self.section2a = strip_section(infile[i+1:],titles)
            if infile[i] == titles[11]:
                self.section2b = strip_section(infile[i+1:],titles)
            if infile[i] == titles[12]:
                self.section2c = strip_section(infile[i+1:],titles)
            if infile[i] == titles[13]:
                self.section2d = strip_section(infile[i+1:],titles)
            # SECTION 3. FREEDOM TO PARTICIPATE IN THE POLITICAL PROCESS
            if infile[i] == titles[14]:
                self.section3 = strip_section(infile[i+1:],titles)
            # SECTION 4. CORRUPTION AND LACK OF TRANSPARENCY IN GOVERNMENT
            if infile[i] == titles[15]:
                self.section4 = strip_section(infile[i+1:],titles)
            # SECTION 5. GOVERNMENTAL ATTITUDE REGARDING INTERNATIONAL
            #  AND NONGOVERNMENTAL INVESTIGATION OF ALLEGED ABUSES OF HUMAN RIGHTS
            if infile[i] == titles[16]:
                self.section5 = strip_section(infile[i+1:],titles)
            # SECTION 6. DISCRIMINATION, SOCIETAL ABUSES, AND TRAFFICKING IN PERSONS
            if infile[i] == 'SECTION 6. DISCRIMINATION, SOCIETAL ABUSES, AND TRAFFICKING IN PERSONS':
                self.section6 = strip_section(infile[i+1:],titles)
            # SECTION 7. WORKER RIGHTS
            if infile[i] == 'a. Freedom of Association and the Right to Collective Bargaining':
                self.section7a = strip_section(infile[i+1:],titles)
            if infile[i] == 'b. Prohibition of Forced or Compulsory Labor':
                self.section7b = strip_section(infile[i+1:],titles)
            if infile[i] == 'c. Prohibition of Child Labor and Minimum Age for Employment':
                self.section7c = strip_section(infile[i+1:],titles)
            if infile[i] == 'd. Discrimination with Respect to Employment and Occupation':
                self.section7d = strip_section(infile[i+1:],titles)
            if infile[i] == 'e. Acceptable Conditions of Work':
                self.section7e = strip_section(infile[i+1:],titles)
        
    def todict(self):
        return {'CountryName' : self.country,
                'Summary' : self.summary,
                'Section_1a' : self.section1a,
                'Section_1b' : self.section1b,
                'Section_1c' : self.section1c,
                'Section_1d' : self.section1d,
                'Section_1e' : self.section1e,
                'Section_1f' : self.section1f,
                'Section_1g' : self.section1g,
                'Section_2a' : self.section2a,
                'Section_2b' : self.section2b,
                'Section_2c' : self.section2c,
                'Section_2d' : self.section2d,
                'Section_3'  : self.section3,
                'Section_4'  : self.section4,
                'Section_5'  : self.section5,
                'Section_6'  : self.section6,
                'Section_7a' : self.section7a,
                'Section_7b' : self.section7b,
                'Section_7c' : self.section7c,
                'Section_7d' : self.section7d,
                'Section_7e' : self.section7e
                }
    
```

# Open and parce file!

```python
f = open("../data/Custom_State_Dep_Reports/2018_all_sections_africa.txt","r+")
master_file = f.read()
```

```python
master_split = master_file.split('\n')
sp_index = [i for i, w in enumerate(master_split) if w == 'EXECUTIVE SUMMARY']
split_index = sp_index + [len(master_split)+1]
```

```python
for j in range(len(split_index)-1):
    parced_data = Parcedatafile(master_split[split_index[j]-1:split_index[j+1]-1],countries)
    State_reports = State_reports.append(parced_data.todict(), ignore_index=True)
```
