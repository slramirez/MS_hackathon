```python
from selenium import webdriver
driver = webdriver.Chrome()
driver.get(url)
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
year_urls = ['http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2012&dlid=204393',
            'http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2013&dlid=220386',
            'http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2014&dlid=236632',    
            'http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2015&dlid=252957',
            'http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2016&dlid=265530',
            'http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2017&dlid=277275',
            'http://www.state.gov/j/drl/rls/hrrpt/humanrightsreport/index.htm?year=2018&dlid=289240']
```

```python
import datetime

print(datetime.datetime.now())

unwanted = ['Preface', 'Overview and Acknowledgements','Appendix A: Notes on Preparation of the Country Reports and Explanatory Material', 'Appendix B: Reporting on Worker Rights', 'Appendix C: Additional Resources', 'Appendix F: United Nations Universal Declaration of Human Rights', 'Appendix G: Errata']
years_list = []

for year in years:
    countries_dict = {}
    path = "//*[contains(@href,\'//www.state.gov/j/drl/rls/hrrpt/{0}/\')]".format(year)
    countries = driver.find_elements_by_xpath(path)
    
    for country in countries:
        countries_dict[country.text] = {'url': country.get_attribute('href')}
        
    for each in unwanted:
        try:
            countries_dict.pop(each)
        except:
            None
        
    for key in countries_dict.keys():
    
        url = countries_dict[key]['url']
        driver.get(url)
    
        headers, subsections = [], []

        for i in range(1,8):
            headers_string = '//*[contains(@class,\'section{0}\')]/h2'.format(i)
            subsections_string = "//*[contains(@class,\'section{0}\')]/div[\'section_content\']".format(str(i))    
    
            headers_list = driver.find_elements_by_xpath(xpath = headers_string)
            num_headers = len(headers_list)
    
            for i in range(num_headers):
                headers.append(headers_list[i].text)
        
            subsections_list = driver.find_elements_by_xpath(xpath = subsections_string)
            num_subsections = len(subsections_list)
    
            if num_subsections == num_headers:
                for i in range(num_subsections):
                    subsections.append(subsections_list[i].text)
            else:
                for i in range(1, num_subsections):
                    subsections.append(subsections_list[i].text)
                
        countries_dict[key] = {'headers': headers,
                              'subsections': subsections} 
        
    years_list.append(countries_dict)
    
print(datetime.datetime.now())
```
