```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df_master=pd.DataFrame(columns = ['Country','Text'])
```

```python

def get_region_links(year):
    
    year_site = requests.get('https://www.state.gov/j/drl/rls/hrrpt/'+str(year)+'/index.htm')
    data = year_site.text
    soup = BeautifulSoup(data)

    region_links = []
    regions = []
    
    for link in soup.find('ul',{"class": "menu"}).find_all('a'):
        region_links.append(link.get('href'))
        regions.append(link.text)
   
    
    ignore = ['Front Matter', 'Appendices', 'Related Material','Appendixes']

    region_tuples = list(zip(region_links, regions))

    new_list = [ x for x in region_tuples if x[1] not in ignore ]
    
    region_links, regions =  zip(*new_list)
    return (region_links, regions)
    
```

```python
import numpy as np
region_codes = ['af','eap','eur','wha','nea','sa']
base_url = 'https://www.state.gov/j/drl/rls/hrrpt/'
years = np.arange(1999,2010,1)
years
```

```python
df_master = pd.DataFrame(columns = ['Country','Text','Year'])

for year in years:
    
    region_links, regions = get_region_links(year)
    
    print('Getting year: ' + str(year))
    
    for i in range(len(region_links)):
                   
        print(regions[i])
        region_site  = requests.get(region_links[i])
        data =region_site.text
        soup = BeautifulSoup(data)

        country_links=[]
        countries=[]

        for link in soup.p.find_all('a',{"target": "_self"}):
            country_links.append(link.get('href'))
            countries.append(link.text)
        
        #print(links)
        #print(countries)


        base_url_2 = 'https://www.state.gov'
        df_reports = pd.DataFrame(columns = ['Country','Text','Year'])


        for i in range(len(country_links)):
            url = base_url_2+country_links[i]
            country = countries[i]
            country_report = requests.get(url)
            data = country_report.text
            soup = BeautifulSoup(data)
            data = soup.find("div", {"id": "centerblock"}).text
            df_reports.loc[i,'Country'] = country
            df_reports.loc[i,'Text'] = data
            df_reports.loc[i,'Year'] = year

        df_master = df_master.append(df_reports)
```
