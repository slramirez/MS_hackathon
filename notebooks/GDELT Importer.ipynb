```python
fp = "../data/20190501.export.CSV"
```

```python
fp_scheme = "../data/gdeltv2_schema.txt"
```

```python
defs = []
desc = []
with open(fp_scheme, 'r') as f:
    for n, line in enumerate(f):
        #print(line.split())
        if n%2 == 0:
            defs.append(line.split())
        else:
            desc.append(line)
```

```python
data = []
for fs, sc in zip(defs, desc):
    data.append(fs + [sc])
```

```python
df_def = pd.DataFrame(data, columns=['name', 'dtype', 'null', 'description'])
```

```python
df_def = df_def[~df_def.name.str.contains('Geo_ADM2Code')]
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def gdelt_page(url):
    """"Scrape news report text from a listing on the GDELT database

    https://blog.gdeltproject.org/gdelt-2-0-our-global-world-in-realtime/

    Parameters
    ----------
    url : str, url of link to report section containing report text

    Returns
    -------
    str, text of report
    """

    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data, 'html.parser')
    return soup
```
